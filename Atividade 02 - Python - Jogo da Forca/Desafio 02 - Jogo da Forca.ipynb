```python
import random
import getpass

# Variáveis
acertos = set()
tentativas = set()
vidas = 7
jogadores = []

# Inserir a palavra e a categoria a ser jogada.'
while True:
  palavra = getpass.getpass("Digite a palavra para ser adivinhada: ")
  if palavra.isalpha() == False:
    limpar_output()
    print("Não é permitido caracteres especiais, números nem palavras compostas. Utilize somente letras. Por favor:") 
  else:
    break
categoria = input("Digite a categoria da palavra: ").upper().strip()

# Inserir 5 jogadores únicos.
while len(jogadores) < 5:
  limpar_output()
  jogador = input('Digite o nome de um jogador: ').upper().strip()
  if jogador == '':
    limpar_output()
    print('Não é permitido espaços em branco. Por favor:')
  elif jogador in jogadores:
      limpar_output()
      print('Este jogador já está inscrito. Por favor:')
  elif jogador.isalpha() == False:
      print("Não é permitido caracteres especiais, números nem espaços. Somente letras. Por favor:") 
  else:
      jogadores.append(jogador)
    

# Escolhe aleatoriamente um jogador.
player = random.choice(jogadores)
limpar_output()
print(f'O jogador escolhido para esta rodada foi: {player}')

# Início do jogo.
while True:
    # Exibir a palavra atual, as vidas e as tentativas restantes.
    bonecos_vida(vidas)
    for letra in palavra:
      if letra in acertos:
        print(letra, end=" ")
      else:
        print("_", end=" ")
    
    print(f'\nA categoria da palavra é: {categoria}')
    print(f"Tentativas restantes: {vidas}")
    

    # Solicita uma letra para o jogador.
    letra = input("Digite uma letra: ").lower()
    
    # Verifica se a letra já foi tentada antes.
    if letra.isalpha() == False:
        limpar_output()
        print('Permitido somente letras.')
    elif len(letra) > 1:
        limpar_output()
        print('Digite apenas uma letra.')
    elif letra in tentativas:
        limpar_output()
        print("Você já tentou usar essa letra.")
    else:
        tentativas.add(letra)
        
        # Verificar se a letra está contida na palavra.
        if letra in palavra:
            acertos.add(letra)
            limpar_output()
            print("Boaa! Acertou!")
            
            # Verifica se o jogador venceu.
            if acertos == set(palavra):
                limpar_output()
                bonecos_vida(vidas)
                print(f"Parabéns {player}, você venceu!")
                print(f"A palavra escolhida foi: {palavra}")
                break
        else:
            vidas = vidas - 1
            limpar_output()
            print("Errou!")
        
        # Verifica se o jogador perdeu.
            if vidas == 0:
                limpar_output()
                bonecos_vida(vidas)
                print(f"Game over {player}! Você perdeu!")
                print(f"A palavra escolhida foi: {palavra}")
                break
```
